```python
import pandas as pd
snowman_return_df = pd.DataFrame(resultData_snowman["Snowman_Return"][1:])
Paytm_Return_df = pd.DataFrame(resultData_paytm["Paytm_Return"][1:])
Zomato_Return_df = pd.DataFrame(resultData_zomato["Zomato_Return"][1:])

combined_df = pd.concat([snowman_return_df, Paytm_Return_df, Zomato_Return_df], axis=1)
combined_df.columns = ["Snowman_Return", "Paytm_Return", "Zomato_Return"]

print(combined_df.head())
combined_df.head()
```

```python
combined_df = combined_df.dropna()
combined_df.head()
```

```python

import pandas as pd
from scipy.stats import levene
df = pd.DataFrame(combined_df)
stat, pval = levene(df['Snowman_Return'], df['Paytm_Return'], df['Zomato_Return'])
print(f'Levene statistic: {stat:.4f}, p-value: {pval:.4f}')
```

```python

import pandas as pd
from scipy import stats

# Load the data from the combined_df dataframe
df = pd.DataFrame(combined_df)

# Extract the three return columns
snowman_return = df['Snowman_Return']
paytm_return = df['Paytm_Return']
zomato_return = df['Zomato_Return']

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(snowman_return, paytm_return, zomato_return)

# Print the results
print(f'F-statistic: {f_statistic:.4f}, p-value: {p_value:.4f}')

# Interpret the results
if p_value < 0.05:
    print('There is a statistically significant difference between the mean returns of the three stocks.')
else:
    print('There is no statistically significant difference between the mean returns of the three stocks.')
```

```python
import pandas as pd
import scipy.stats as stats

# Load the data from the combined_df dataframe
df = pd.DataFrame(combined_df)

# Extract the three return columns
snowman_return = df['Snowman_Return']
paytm_return = df['Paytm_Return']
zomato_return = df['Zomato_Return']

# Perform two-sample t-tests between each pair of stocks
t_statistic, p_value = stats.ttest_ind(snowman_return, paytm_return)
print(f'snowman and zomato T-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}')

t_statistic, p_value = stats.ttest_ind(snowman_return, zomato_return)
print(f'T-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}')

t_statistic, p_value = stats.ttest_ind(paytm_return, zomato_return)
print(f'T-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}')

t_statistic, p_value = stats.ttest_ind(paytm_return, snowman_return)
print(f'T-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}')

t_statistic, p_value = stats.ttest_ind(zomato_return, snowman_return)
print(f'T-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}')

t_statistic, p_value = stats.ttest_ind(zomato_return, paytm_return)
print(f'T-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}')
```

```python
####Q3# Finding beta of all the 3 companies
# downloading data of snowman logistics from 05 feb 2023 to 05 feb 2024
import yfinance as yf # load yahoo finance as yf
import pandas as pd

# setting the ticker value
ticker = 'SNOWMAN.ns'
# giving the start and end dates
startDate = '2023-02-05'
endDate = '2024-02-05'

# downloading the data of the ticker value between
# the start and end dates
resultData_snowman = yf.download(ticker, startDate, endDate) # call the function to get data

# printing the last 5 rows of the data
print(resultData_snowman)
resultData_snowman.head()
```

```python
resultData_snowman["Snowman_Return"] = resultData_snowman["Adj Close"].pct_change() # percentage change
resultData_snowman["Snowman_Return"] # (Today Price - Yesterday Price) /Yesterday Price
resultData_snowman["Snowman_Return"][1:]
```

```python
ticker = "^NSEI"
df_nifty = yf.download(ticker, start="2023-02-05", end="2024-02-05")
df_nifty.head()
```

```python
df_nifty["NIFTY_50"] = df_nifty["Adj Close"].pct_change()
df_nifty["NIFTY_50"][1:]
```

```python
# Beta of a Stock
y = df_nifty["NIFTY_50"][1:]
x = resultData_snowman["Snowman_Return"][1:]
# Beta = measure of risk;  higher the risk; higher the return
# Beta formula from stats Beta = Covar(Returns of Reliance, Returns of NIFTY)/ Variance of Nifty
# Python does not have a direct formula for computing Beta
# Indirect: Correlation = Cov(Reliance_Ret, NIFTY_Ret)/ Std.dev(Reliance). Std.dev(NIFTY)
correlation = x.corr(y)
# Standard Deviation of Reliance and Standard Deviation of NIFTY
xstd = x.std()
ystd = y.std()
# Correltion * Std. Dev Nifty / Std.Dev RIL = Covar(Reliance, NIFTY)/ (std.NIFTY, std.Rel)
Beta = correlation * xstd/ystd
Beta
```
